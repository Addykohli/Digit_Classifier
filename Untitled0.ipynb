```python
import tensorflow as tf
import numpy as np
import random
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import layers
import os
import math
from google.colab import files
```

```python


# Path to dataset
train_dir = "/content/drive/My Drive/NumClassifyData"

# Image parameters
IMG_SIZE = (90, 140)  # Resize all images
BATCH_SIZE = 32
ROTATION_RANGE = 30  # Rotate images between -30° to +30°
AUGMENT_COUNT = 300  # Number of augmented images per class

# Load training data
train_ds = tf.keras.preprocessing.image_dataset_from_directory(
    train_dir,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    label_mode="int",
    validation_split=0.2,
    subset="training",
    seed=123
)

val_ds = tf.keras.preprocessing.image_dataset_from_directory(
    train_dir,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    label_mode="int",
    validation_split=0.2,
    subset="validation",
    seed=123
)

# Function to apply random rotation
def augment_images(images, labels):
    datagen = ImageDataGenerator(rotation_range=ROTATION_RANGE)
    augmented_images = []
    augmented_labels = []

    for img, label in zip(images, labels):
        img = np.expand_dims(img, axis=0)  # Expand dims for batch processing

        i = 0
        for batch in datagen.flow(img, batch_size=1):
            augmented_images.append(batch[0])  # Append rotated image
            augmented_labels.append(label)  # Same label
            i += 1
            if i >= AUGMENT_COUNT // len(images):  # Generate required images
                break

    return np.array(augmented_images), np.array(augmented_labels)

# Convert train_ds to numpy arrays
train_images = []
train_labels = []

for images, labels in train_ds:
    train_images.append(images.numpy())
    train_labels.append(labels.numpy())

train_images = np.concatenate(train_images)
train_labels = np.concatenate(train_labels)

# Apply augmentation and add new images to dataset
aug_images, aug_labels = augment_images(train_images, train_labels)
train_images = np.concatenate([train_images, aug_images])
train_labels = np.concatenate([train_labels, aug_labels])

# Convert back to TensorFlow dataset
train_ds = tf.data.Dataset.from_tensor_slices((train_images / 255.0, train_labels)).batch(BATCH_SIZE)
val_ds = val_ds.map(lambda x, y: (x / 255.0, y))  # Normalize validation set

# Define the CNN model
model = tf.keras.Sequential([
    layers.InputLayer(shape=(90, 140, 3)),  # Define input shape here
    layers.Conv2D(32, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),

    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),

    layers.Conv2D(128, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),

    layers.Flatten(),
    layers.Dense(256, activation='relu'),
    layers.Dense(10, activation='softmax')  # 10 output classes (0-9)
])

# Compile the model
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Custom callback to print epoch and accuracy
class PrintEpochAccuracy(tf.keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs=None):
        val_accuracy = logs.get('val_accuracy', None)
        if val_accuracy is not None:
            print(f"Epoch {epoch + 1}: Validation Accuracy = {val_accuracy:.4f}")
        else:
            print(f"Epoch {epoch + 1}: Validation Accuracy = N/A")

# Train the model
model.fit(train_ds,
          epochs=20,
          validation_data=val_ds,
          callbacks=[PrintEpochAccuracy()])
```

```python
test_folders = [
    "/content/drive/MyDrive/example_test1",
    "/content/drive/MyDrive/example_test2",
    "/content/drive/MyDrive/example_test3",
    "/content/drive/MyDrive/example_test4"
]

# Function to load and preprocess images from test folders
def load_and_preprocess_images(test_folders, img_size=(90, 140)):
    images = []
    filenames = []
    img_paths = []
    for folder in test_folders:
        for filename in os.listdir(folder):
            img_path = os.path.join(folder, filename)
            img = image.load_img(img_path, target_size=img_size)
            img_array = image.img_to_array(img)
            images.append(img_array)
            filenames.append(filename)
            img_paths.append(img_path)
    images = np.array(images) / 255.0  # Normalize images to [0, 1]
    return images, filenames, img_paths

# Load images
test_images, test_filenames, test_img_paths = load_and_preprocess_images(test_folders)

# Make predictions
predictions = model.predict(test_images)
predicted_labels = np.argmax(predictions, axis=1)

# Determine grid size
num_images = len(test_filenames)
cols = 8  # Set number of columns
rows = math.ceil(num_images / cols)  # Calculate required rows

# Display images with predictions
fig, axes = plt.subplots(rows, cols, figsize=(15, rows * 2))

for i, (filename, predicted_label, img_path) in enumerate(zip(test_filenames, predicted_labels, test_img_paths)):
    img = image.load_img(img_path)

    ax = axes[i // cols, i % cols]  # Position in grid
    ax.imshow(img)
    ax.set_title(f"Pred: {predicted_label}", fontsize=8)
    ax.axis("off")

# Hide any empty subplots
for j in range(i + 1, rows * cols):
    axes[j // cols, j % cols].axis("off")

plt.tight_layout()
plt.show()

```

```python
# Define the test folder
test_folders = [
    "/content/drive/MyDrive/digital_digits_test",
]

# Function to load and preprocess images
def load_and_preprocess_images(test_folders, img_size=(90, 140)):
    images = {}
    for folder in test_folders:
        for filename in os.listdir(folder):
            if filename.endswith(".png") or filename.endswith(".jpg"):
                digit = int(filename.split('.')[0])  # Extract the digit from the filename
                img_path = os.path.join(folder, filename)
                img = image.load_img(img_path, target_size=img_size)
                img_array = image.img_to_array(img)
                images[digit] = img_array / 255.0  # Normalize images to [0, 1]
    return images

# Load images
test_images = load_and_preprocess_images(test_folders)

# Define the display order
image_order = [
    [8, 3],
    [2, 7, 1],
    [6, 5],
    [9, 4],
    [0]
]

# Display images
fig, axes = plt.subplots(len(image_order), max(map(len, image_order)), figsize=(10, 8))

for row_idx, row in enumerate(image_order):
    for col_idx, digit in enumerate(row):
        ax = axes[row_idx, col_idx]
        ax.imshow(test_images[digit])
        ax.set_title(f"{digit}")
        ax.axis("off")

# Hide empty subplots
for row_idx in range(len(image_order)):
    for col_idx in range(len(image_order[row_idx]), max(map(len, image_order))):
        axes[row_idx, col_idx].axis("off")

plt.tight_layout()
plt.show()
```
